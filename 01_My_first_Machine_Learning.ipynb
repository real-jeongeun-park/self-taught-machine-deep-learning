# 01-1. 인공지능과 머신러닝, *딥러닝*

### 인공지능

- **인공지능**이란? 사람처럼 학습하고 추론할 수 있는 지능을 가진 컴퓨터 시스템을 만드는 기술 또는 분야.
- 인공지능의 분류
    - **인공일반지능 또는 강인공지능(Strong AI)**: 사람과 구분하기 어려운 지능을 가진 컴퓨터 시스템
    - **약인공지능(Weak AI**): 특정 분야에서 사람의 일을 보조하는, 지능을 가진 컴퓨터 시스템

### 머신러닝

- **머신러닝**이란? 규칙을 프로그래밍 하지 않고도 자동으로 데이터로 규칙을 학습하는 알고리즘을 연구하는 분야. 결국에는 알고리즘임
- 컴퓨터 과학 분야의 대표적인 머신러닝 라이브러리로 **사이킷런**(scikit-learn)이 있음

### 딥러닝

- **딥러닝**이란? 머신러닝 알고리즘 중에서도 인공 신경망(Artificial neural network)을 기반으로 한 알고리즘을 연구하는 분야임.
- 유명한 딥러닝 라이브러리로 구글의 **텐서플로**(Tensorflow)와 메타의 **파이토치**(PyTorch)가 있음

# 01-2 코랩과 주피터 노트북

### 도미 데이터 준비하기

- 도미와 빙어를 구분하는 이진 분류를 진행한다.
- **이진 분류(binary classification)**란 두 개의 클래스(종류) 중 하나를 고르는 문제를 뜻한다. 파이썬 프로그램의 클래스와는 다른 개념이다.
- bream_length와 bream_weight를 특성(feature)라 한다. **특성은 데이터의 특징으로 데이터 표의 각 컬럼을 의미한다.**

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

print(bream_length)
print(bream_weight)
```

- 참고
    - **feature**: 입력값. 예측에 쓰이는 변수. 하나의 열.
    - **sample**: 한 행. 데이터 하나.
    - **label**: 정답. 예측하고 싶은 값. 예시: 물고기의 종류
    - **dataset**: 전체 데이터 테이블.
- 데이터를 더 잘 이해하기 위해 산점도로 표현한다. **산점도**(Scatter plot)란 x,y축으로 이루어진 좌표계에 두 변수(x, y)의 관계를 점으로 표현한 그림이다.

```python
import matplotlib.pyplot as plt
plt.scatter(bream_length, bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
import matplotlib.pyplot as plt

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### 도미와 빙어 데이터 매핑하기

- **사이킷런**의 k-최근접 이웃 알고리즘을 사용하여 도미와 빙어를 구분한다. 사이킷런에서는 input으로 반드시 2차원 리스트를 사용한다.
- 아래의 코드에서는 zip과 list comprehension 구문을 사용하여 쉽게 데이터셋인 fish_data를 만든다. 각 low는 length, weight로 이루어져 있다.

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight  # 순서로 도미, 빙어

fish_data = [[l,w] for l, w in zip(length, weight)]
print(fish_data)
```

### 정답 데이터 준비하기

- 도미를 숫자 1, 빙어를 0으로 표현한 리스트를 생성한다. 이는 **이진 분류**에서 찾으려는 대상을 1로 놓고, 그 외에는 0으로 놓기 때문이다. 참고로 1에 해당하는 클래스를 양성 클래스(positve class), 0에 해당하는 클래스를 음성 클래스(negative class)라 한다.
- 아래 코드에서 `fish_target` 리스트가 있다. **target**이란 label과 같은 의미로 예측하고 싶은 값, 즉 **정답**을 의미한다.

```python
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

### k-최근접 이웃 알고리즘 사용하기

- 사이킷런 패키지에서 k-최근접 알고리즘을 구현한 클래스인 `KNeighborsClassifier`를 임포트한다.:

```python
from sklearn.neighbors import KNeighborsClassifier
# import sklearn
# model = sklearn.neigbors.KNeighborsClassifier() 와 같음
```

- `fit` 메소드에 fish_data(데이터셋)과 fish_target(정답 리스트)를 전달하여 도미를 찾기 위한 **규칙을 학습시킨다.** 이런 과정을 머신러닝에서는 **훈련(training)**이라고 한다.
- `score` 메소드에 똑같은 input을 주고, 객체(모델)이 얼마나 잘 훈련됐는지를 평가한다. score 메소드는 0에서 1 사이의 값을 반환한다. 1은 모든 데이터를 정확하게 맞혔다는 의미이고, 0은 어떤 데이터도 맞히지 못했다는 의미다.

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

- 여기서 **모델**이란 머신러닝 알고리즘으로 구현한 프로그램이다. 위의 경우 kn이 그 예다.
- **k-최근접 이웃 알고리즘**
    - 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는 것을 정답으로 사용한다. 다수결의 원칙이라고 생각하면 쉽다.
    - 새로운 데이터에 대해 예측할 때 가장 가까운 직선 거리에 어떤 데이터가 많은지만을 살피면 된다. 따라서, 데이터가 아주 많은 경우 계산량이 크게 증가한다는 단점이 있다.
- 예를 들어 아래의 초록색 점의 경우 k-최근접 이웃 알고리즘은 도미라고 판단할 것이다. 주위의 데이터가 모두 도미에 해당하기 때문이다.

predict 메소드는 새로운 데이터에 대한 정답을 예측한다. 길이가 30이고, 무게가 600인 물고기가 무엇인지 예측하고 싶다면 아래와 같이 코드를 작성한다. 중요한 것은, 반드시 input은 2차원 리스트로 들어가야 한다는 것이다. 결과는 array([1])로, 도미로 예측한다.

```python
kn.predict([[30,600]])
```

k-최근접 이웃 알고리즘은 기본적으로 가장 가까운 5개의 데이터를 참고한다. 원하면 다른 숫자로 바꿀 수도 있다. 아래 코드의 경우 가장 가까운 49개의 데이터를 참고한다. 그런데 이렇게 되면 모델은 모든 생선 데이터를 참고하게 되고, 결국 데이터 중 다수를 차지하는 도미가 어떤 경우에도 무조건 정답으로 나오게 된다.

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target) # 훈련 시킴
kn49.score(fish_data, fish_target) ## 대략 0.7 = 35/49임 (생선 중 도미 데이터 수)
```

# 전체 코드

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter([30], [600])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

length = bream_length + smelt_length
weight = bream_weight + smelt_weight  # 순서로 도미, 빙어

fish_data = [[l,w] for l, w in zip(length, weight)]
print(fish_data)

fish_target = [1] * 35 + [0] * 14
print(fish_target)

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)

kn.predict([[30,600]])

kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)

print(kn49.score(fish_data, fish_target))
print(35/49)
```
