### ÎèÑÎØ∏ Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑÌïòÍ∏∞

- ÎèÑÎØ∏ÏôÄ ÎπôÏñ¥Î•º Íµ¨Î∂ÑÌïòÎäî Ïù¥ÏßÑ Î∂ÑÎ•òÎ•º ÏßÑÌñâÌïúÎã§.
- **Ïù¥ÏßÑ Î∂ÑÎ•ò(binary classification)**ÎûÄ Îëê Í∞úÏùò ÌÅ¥ÎûòÏä§(Ï¢ÖÎ•ò) Ï§ë ÌïòÎÇòÎ•º Í≥†Î•¥Îäî Î¨∏Ï†úÎ•º ÎúªÌïúÎã§. ÌååÏù¥Ïç¨ ÌîÑÎ°úÍ∑∏Îû®Ïùò ÌÅ¥ÎûòÏä§ÏôÄÎäî Îã§Î•∏ Í∞úÎÖêÏù¥Îã§.
- bream_lengthÏôÄ bream_weightÎ•º ÌäπÏÑ±(feature)Îùº ÌïúÎã§. **ÌäπÏÑ±ÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò ÌäπÏßïÏúºÎ°ú Îç∞Ïù¥ÌÑ∞ ÌëúÏùò Í∞Å Ïª¨ÎüºÏùÑ ÏùòÎØ∏ÌïúÎã§.**

```python
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

print(bream_length)
print(bream_weight)
```

- Ï∞∏Í≥†
    - **feature**: ÏûÖÎ†•Í∞í. ÏòàÏ∏°Ïóê Ïì∞Ïù¥Îäî Î≥ÄÏàò. ÌïòÎÇòÏùò Ïó¥.
    - **sample**: Ìïú Ìñâ. Îç∞Ïù¥ÌÑ∞ ÌïòÎÇò.
    - **label**: Ï†ïÎãµ. ÏòàÏ∏°ÌïòÍ≥† Ïã∂ÏùÄ Í∞í. ÏòàÏãú: Î¨ºÍ≥†Í∏∞Ïùò Ï¢ÖÎ•ò
    - **dataset**: Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÌÖåÏù¥Î∏î.
- Îç∞Ïù¥ÌÑ∞Î•º Îçî Ïûò Ïù¥Ìï¥ÌïòÍ∏∞ ÏúÑÌï¥ ÏÇ∞Ï†êÎèÑÎ°ú ÌëúÌòÑÌïúÎã§. **ÏÇ∞Ï†êÎèÑ**(Scatter plot)ÎûÄ x,yÏ∂ïÏúºÎ°ú Ïù¥Î£®Ïñ¥ÏßÑ Ï¢åÌëúÍ≥ÑÏóê Îëê Î≥ÄÏàò(x, y)Ïùò Í¥ÄÍ≥ÑÎ•º Ï†êÏúºÎ°ú ÌëúÌòÑÌïú Í∑∏Î¶ºÏù¥Îã§.

```python
import matplotlib.pyplot as plt

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

### ÎèÑÎØ∏ÏôÄ ÎπôÏñ¥ Îç∞Ïù¥ÌÑ∞ Îß§ÌïëÌïòÍ∏∞

- **ÏÇ¨Ïù¥ÌÇ∑Îü∞**Ïùò k-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶òÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÎèÑÎØ∏ÏôÄ ÎπôÏñ¥Î•º Íµ¨Î∂ÑÌïúÎã§. ÏÇ¨Ïù¥ÌÇ∑Îü∞ÏóêÏÑúÎäî inputÏúºÎ°ú Î∞òÎìúÏãú 2Ï∞®Ïõê Î¶¨Ïä§Ìä∏Î•º ÏÇ¨Ïö©ÌïúÎã§.
- ÏïÑÎûòÏùò ÏΩîÎìúÏóêÏÑúÎäî zipÍ≥º list comprehension Íµ¨Î¨∏ÏùÑ ÏÇ¨Ïö©ÌïòÏó¨ ÏâΩÍ≤å Îç∞Ïù¥ÌÑ∞ÏÖãÏù∏ fish_dataÎ•º ÎßåÎì†Îã§. Í∞Å lowÎäî length, weightÎ°ú Ïù¥Î£®Ïñ¥Ï†∏ ÏûàÎã§.

```python
length = bream_length + smelt_length
weight = bream_weight + smelt_weight  # ÏàúÏÑúÎ°ú ÎèÑÎØ∏, ÎπôÏñ¥

fish_data = [[l,w] for l, w in zip(length, weight)]
print(fish_data)
```

### Ï†ïÎãµ Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑÌïòÍ∏∞

- ÎèÑÎØ∏Î•º Ïà´Ïûê 1, ÎπôÏñ¥Î•º 0ÏúºÎ°ú ÌëúÌòÑÌïú Î¶¨Ïä§Ìä∏Î•º ÏÉùÏÑ±ÌïúÎã§. Ïù¥Îäî **Ïù¥ÏßÑ Î∂ÑÎ•ò**ÏóêÏÑú Ï∞æÏúºÎ†§Îäî ÎåÄÏÉÅÏùÑ 1Î°ú ÎÜìÍ≥†, Í∑∏ Ïô∏ÏóêÎäî 0ÏúºÎ°ú ÎÜìÍ∏∞ ÎïåÎ¨∏Ïù¥Îã§. Ï∞∏Í≥†Î°ú 1Ïóê Ìï¥ÎãπÌïòÎäî ÌÅ¥ÎûòÏä§Î•º ÏñëÏÑ± ÌÅ¥ÎûòÏä§(positve class), 0Ïóê Ìï¥ÎãπÌïòÎäî ÌÅ¥ÎûòÏä§Î•º ÏùåÏÑ± ÌÅ¥ÎûòÏä§(negative class)Îùº ÌïúÎã§.
- ÏïÑÎûò ÏΩîÎìúÏóêÏÑú `fish_target` Î¶¨Ïä§Ìä∏Í∞Ä ÏûàÎã§. **target**Ïù¥ÎûÄ labelÍ≥º Í∞ôÏùÄ ÏùòÎØ∏Î°ú ÏòàÏ∏°ÌïòÍ≥† Ïã∂ÏùÄ Í∞í, Ï¶â **Ï†ïÎãµ**ÏùÑ ÏùòÎØ∏ÌïúÎã§.

```python
fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

### k-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶ò ÏÇ¨Ïö©ÌïòÍ∏∞

- ÏÇ¨Ïù¥ÌÇ∑Îü∞ Ìå®ÌÇ§ÏßÄÏóêÏÑú k-ÏµúÍ∑ºÏ†ë ÏïåÍ≥†Î¶¨Ï¶òÏùÑ Íµ¨ÌòÑÌïú ÌÅ¥ÎûòÏä§Ïù∏ `KNeighborsClassifier`Î•º ÏûÑÌè¨Ìä∏ÌïúÎã§.:

```python
from sklearn.neighbors import KNeighborsClassifier
# import sklearn
# model = sklearn.neigbors.KNeighborsClassifier() ÏôÄ Í∞ôÏùå
```

- `fit` Î©îÏÜåÎìúÏóê fish_data(Îç∞Ïù¥ÌÑ∞ÏÖã)Í≥º fish_target(Ï†ïÎãµ Î¶¨Ïä§Ìä∏)Î•º Ï†ÑÎã¨ÌïòÏó¨ ÎèÑÎØ∏Î•º Ï∞æÍ∏∞ ÏúÑÌïú **Í∑úÏπôÏùÑ ÌïôÏäµÏãúÌÇ®Îã§.** Ïù¥Îü∞ Í≥ºÏ†ïÏùÑ Î®∏Ïã†Îü¨ÎãùÏóêÏÑúÎäî **ÌõàÎ†®(training)**Ïù¥ÎùºÍ≥† ÌïúÎã§.
- `score` Î©îÏÜåÎìúÏóê ÎòëÍ∞ôÏùÄ inputÏùÑ Ï£ºÍ≥†, Í∞ùÏ≤¥(Î™®Îç∏)Ïù¥ ÏñºÎßàÎÇò Ïûò ÌõàÎ†®ÎêêÎäîÏßÄÎ•º ÌèâÍ∞ÄÌïúÎã§. score Î©îÏÜåÎìúÎäî 0ÏóêÏÑú 1 ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ Î∞òÌôòÌïúÎã§. 1ÏùÄ Î™®Îì† Îç∞Ïù¥ÌÑ∞Î•º Ï†ïÌôïÌïòÍ≤å ÎßûÌòîÎã§Îäî ÏùòÎØ∏Ïù¥Í≥†, 0ÏùÄ Ïñ¥Îñ§ Îç∞Ïù¥ÌÑ∞ÎèÑ ÎßûÌûàÏßÄ Î™ªÌñàÎã§Îäî ÏùòÎØ∏Îã§.

```python
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)
```

- Ïó¨Í∏∞ÏÑú **Î™®Îç∏**Ïù¥ÎûÄ Î®∏Ïã†Îü¨Îãù ÏïåÍ≥†Î¶¨Ï¶òÏúºÎ°ú Íµ¨ÌòÑÌïú ÌîÑÎ°úÍ∑∏Îû®Ïù¥Îã§. ÏúÑÏùò Í≤ΩÏö∞ knÏù¥ Í∑∏ ÏòàÎã§.
- **k-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶ò**
    - Ïñ¥Îñ§ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÎãµÏùÑ Íµ¨Ìï† Îïå Ï£ºÏúÑÏùò Îã§Î•∏ Îç∞Ïù¥ÌÑ∞Î•º Î≥¥Í≥† Îã§ÏàòÎ•º Ï∞®ÏßÄÌïòÎäî Í≤ÉÏùÑ Ï†ïÎãµÏúºÎ°ú ÏÇ¨Ïö©ÌïúÎã§. Îã§ÏàòÍ≤∞Ïùò ÏõêÏπôÏù¥ÎùºÍ≥† ÏÉùÍ∞ÅÌïòÎ©¥ ÏâΩÎã§.
    - ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌï¥ ÏòàÏ∏°Ìï† Îïå Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÏßÅÏÑ† Í±∞Î¶¨Ïóê Ïñ¥Îñ§ Îç∞Ïù¥ÌÑ∞Í∞Ä ÎßéÏùÄÏßÄÎßåÏùÑ ÏÇ¥ÌîºÎ©¥ ÎêúÎã§. Îî∞ÎùºÏÑú, Îç∞Ïù¥ÌÑ∞Í∞Ä ÏïÑÏ£º ÎßéÏùÄ Í≤ΩÏö∞ Í≥ÑÏÇ∞ÎüâÏù¥ ÌÅ¨Í≤å Ï¶ùÍ∞ÄÌïúÎã§Îäî Îã®Ï†êÏù¥ ÏûàÎã§.
- ÏòàÎ•º Îì§Ïñ¥ ÏïÑÎûòÏùò Ï¥àÎ°ùÏÉâ Ï†êÏùò Í≤ΩÏö∞ k-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶òÏùÄ ÎèÑÎØ∏ÎùºÍ≥† ÌåêÎã®Ìï† Í≤ÉÏù¥Îã§. Ï£ºÏúÑÏùò Îç∞Ïù¥ÌÑ∞Í∞Ä Î™®Îëê ÎèÑÎØ∏Ïóê Ìï¥ÎãπÌïòÍ∏∞ ÎïåÎ¨∏Ïù¥Îã§.

predict Î©îÏÜåÎìúÎäî ÏÉàÎ°úÏö¥ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú Ï†ïÎãµÏùÑ ÏòàÏ∏°ÌïúÎã§. Í∏∏Ïù¥Í∞Ä 30Ïù¥Í≥†, Î¨¥Í≤åÍ∞Ä 600Ïù∏ Î¨ºÍ≥†Í∏∞Í∞Ä Î¨¥ÏóáÏù∏ÏßÄ ÏòàÏ∏°ÌïòÍ≥† Ïã∂Îã§Î©¥ ÏïÑÎûòÏôÄ Í∞ôÏù¥ ÏΩîÎìúÎ•º ÏûëÏÑ±ÌïúÎã§. Ï§ëÏöîÌïú Í≤ÉÏùÄ, Î∞òÎìúÏãú inputÏùÄ 2Ï∞®Ïõê Î¶¨Ïä§Ìä∏Î°ú Îì§Ïñ¥Í∞ÄÏïº ÌïúÎã§Îäî Í≤ÉÏù¥Îã§. Í≤∞Í≥ºÎäî array([1])Î°ú, ÎèÑÎØ∏Î°ú ÏòàÏ∏°ÌïúÎã§.

```python
kn.predict([[30,600]])
```

k-ÏµúÍ∑ºÏ†ë Ïù¥ÏõÉ ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Í∏∞Î≥∏Ï†ÅÏúºÎ°ú Í∞ÄÏû• Í∞ÄÍπåÏö¥ 5Í∞úÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∞∏Í≥†ÌïúÎã§. ÏõêÌïòÎ©¥ Îã§Î•∏ Ïà´ÏûêÎ°ú Î∞îÍøÄ ÏàòÎèÑ ÏûàÎã§. ÏïÑÎûò ÏΩîÎìúÏùò Í≤ΩÏö∞ Í∞ÄÏû• Í∞ÄÍπåÏö¥ 49Í∞úÏùò Îç∞Ïù¥ÌÑ∞Î•º Ï∞∏Í≥†ÌïúÎã§. Í∑∏Îü∞Îç∞ Ïù¥Î†áÍ≤å ÎêòÎ©¥ Î™®Îç∏ÏùÄ Î™®Îì† ÏÉùÏÑ† Îç∞Ïù¥ÌÑ∞Î•º Ï∞∏Í≥†ÌïòÍ≤å ÎêòÍ≥†, Í≤∞Íµ≠ Îç∞Ïù¥ÌÑ∞ Ï§ë Îã§ÏàòÎ•º Ï∞®ÏßÄÌïòÎäî ÎèÑÎØ∏Í∞Ä Ïñ¥Îñ§ Í≤ΩÏö∞ÏóêÎèÑ Î¨¥Ï°∞Í±¥ Ï†ïÎãµÏúºÎ°ú ÎÇòÏò§Í≤å ÎêúÎã§.

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target) # ÌõàÎ†® ÏãúÌÇ¥
kn49.score(fish_data, fish_target) ## ÎåÄÎûµ 0.7 = 35/49ÏûÑ (ÏÉùÏÑ† Ï§ë ÎèÑÎØ∏ Îç∞Ïù¥ÌÑ∞ Ïàò)
```

# Ï†ÑÏ≤¥ ÏΩîÎìú

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter([30], [600])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

length = bream_length + smelt_length
weight = bream_weight + smelt_weight  # ÏàúÏÑúÎ°ú ÎèÑÎØ∏, ÎπôÏñ¥

fish_data = [[l,w] for l, w in zip(length, weight)]
print(fish_data)

fish_target = [1] * 35 + [0] * 14
print(fish_target)

kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)
kn.score(fish_data, fish_target)

kn.predict([[30,600]])

kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data, fish_target)

print(kn49.score(fish_data, fish_target))
print(35/49)
```
